## 30.

```python
import re

neko = []

with open('out/neko.txt.mecab') as f:
    lf = False
    sentence = []
    for l in f:
        if l == '\n':
            lf = True
        elif l == 'EOS\n':
            lf = False
            neko.append(sentence)
            sentence = []
        else:
            if lf:
                l = '\n' + l
                lf = False
            l = l.rstrip()
            surface, info_ = l.split('\t')
            info = info_.split(',')
            base = info[4]
            pos = info[0]
            pos1 = info[1]
            sentence.append({
                'surface': surface,
                'base': base,
                'pos': pos,
                'pos1': pos1
            })

print(neko)
```

## 31.

## 33.

```python
for sentence in neko:
    for i in range(len(sentence) - 2):
        morpheme0 = sentence[i]
        morpheme1 = sentence[i + 1]
        morpheme2 = sentence[i + 2]
        if morpheme0['pos'] == '名詞' and morpheme1['base'] == 'の' and morpheme1['pos1'] == '接続助詞' and morpheme2['pos'] == '名詞':
            print(morpheme0['surface'] + morpheme1['surface'] + morpheme2['surface'])
```

## 34.

```python
for sentence in neko:
    n = len(sentence)
    i = 0
    while i < n:
        while i < n and sentence[i]['pos'] != '名詞':
            i += 1
        start = i
        phrase = ''
        while i < n and sentence[i]['pos'] == '名詞':
            phrase += sentence[i]['surface']
            i += 1
        stop = i
        if stop - start >= 2:
            print(phrase)
```

## 35.

```python
tally = dict()
for sentence in neko:
    for morpheme in sentence:
        if morpheme['pos'] == '特殊':
            continue
        base = morpheme['base']
        tally[base] = tally.get(base, 0) + 1

for base, count in reversed(sorted(tally.items(), key=lambda c: c[1])):
    print(count, base)
```

```python
from itertools import islice
import matplotlib.pyplot as plt
import japanize_matplotlib

tally = dict()
for sentence in neko:
    for morpheme in sentence:
        if morpheme['pos'] == '特殊':
            continue
        word = morpheme['surface'] if morpheme['base'] == '*' else morpheme['base']
        tally[word] = tally.get(word, 0) + 1

words = []
counts = []

for word, count in islice(reversed(sorted(tally.items(), key=lambda c: c[1])), 10):
    words.append(word)
    counts.append(count)

plt.bar(range(0, 10), counts, tick_label=words)
None
```

```python
from itertools import islice
import matplotlib.pyplot as plt
import japanize_matplotlib

tally = dict()
for sentence in neko:
    if any(morpheme['base'] == '猫' for morpheme in sentence):
        for morpheme in sentence:
            if morpheme['pos'] == '特殊':
                continue
            if morpheme['base'] == '猫':
                continue
            word = morpheme['surface'] if morpheme['base'] == '*' else morpheme['base']
            tally[word] = tally.get(word, 0) + 1

words = []
counts = []

for word, count in islice(reversed(sorted(tally.items(), key=lambda c: c[1])), 10):
    words.append(word)
    counts.append(count)

plt.bar(range(0, 10), counts, tick_label=words)
None
```

## 38.

```python
import matplotlib.pyplot as plt
import japanize_matplotlib

tally = dict()
for sentence in neko:
    for morpheme in sentence:
        if morpheme['pos'] == '特殊':
            continue
        word = morpheme['surface'] if morpheme['base'] == '*' else morpheme['base']
        tally[word] = tally.get(word, 0) + 1

max_count = max(tally.values())
freq = [0] * (max_count + 1)

for word, count in tally.items():
    freq[count] += 1

ax = plt.figure().add_subplot(xlabel='Rank', ylabel='Frequency')
ax.hist(freq, bins=40, range=(1, max_count))
None
```
