## 30.

```python
import re

neko = []

with open('out/neko.txt.mecab') as f:
    lf = False
    sentence = []
    for l in f:
        if l == '\n':
            lf = True
        elif l == 'EOS\n':
            lf = False
            neko.append(sentence)
            sentence = []
        else:
            if lf:
                l = '\n' + l
                lf = False
            l = l.rstrip()
            surface, info_ = l.split('\t')
            info = info_.split(',')
            base = info[4]
            pos = info[0]
            pos1 = info[1]
            sentence.append({
                'surface': surface,
                'base': base,
                'pos': pos,
                'pos1': pos1
            })

print(neko)
```

## 31.

```python
for sentence in neko:
    for morpheme in sentence:
        if morpheme['pos'] == '動詞':
            print(morpheme['surface'])
```

## 32.

```python
for sentence in neko:
    for morpheme in sentence:
        if morpheme['pos'] == '動詞':
            print(morpheme['base'])
```

## 33.

```python
for sentence in neko:
    for i in range(len(sentence) - 2):
        morpheme0 = sentence[i]
        morpheme1 = sentence[i + 1]
        morpheme2 = sentence[i + 2]
        if morpheme0['pos'] == '名詞' and morpheme1['base'] == 'の' and morpheme1['pos1'] == '接続助詞' and morpheme2['pos'] == '名詞':
            print(morpheme0['surface'] + morpheme1['surface'] + morpheme2['surface'])
```
